```python
#top 10 wicket takers overall

bowlers = list(ipl_df['bowler'].unique())

def wickets_taken(df,name):
    x = ipl_df[(df['player_dismissed'].notna()) &
               (df['bowler'] == name) & 
               (df['wicket_type'] != 'run out') & 
               (df['wicket_type'] != 'retired hurt') & 
               (df['wicket_type'] !='hit wicket') &
               (df['wicket_type'] !='obstructing the field')]
    return x.shape[0]

w = {}
for i in bowlers:
    w[i] = wickets_taken(ipl_df,i)
top_10_wicket_takers_overall = dict(sorted(w.items(), key = lambda x: x[1], reverse = True)[:10])

fig = plt.figure(figsize = (12, 9))
bars = plt.bar(list(top_10_wicket_takers_overall.keys())[::-1],list(top_10_wicket_takers_overall.values())[::-1])
plt.ylabel("Bowlers")
plt.title("top 10 wicket takers overall")
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x(), yval + .005, yval)
plt.show()
```

```python
#top 10 run scoreres

batsmen = list(ipl_df['striker'].unique())

def runs_scored(df,name):
    x = df[(df['player_dismissed'].isna()) &
               (df['striker'] == name) & 
               (df['extras']==0) & 
               (df['noballs'].isna()) & 
               (df['byes'].isna()) & 
               (df['legbyes'].isna()) &
               (df['penalty'].isna()) ]
    
    return x['runs_off_bat'].sum()

r = {}
for i in batsmen:
    r[i]=runs_scored(ipl_df,i)
top_10_run_scoreres = dict(sorted(r.items(), key = lambda x: x[1], reverse = True)[:10])

fig = plt.figure(figsize = (18, 5))
bars = plt.bar(list(top_10_run_scoreres.keys()),list(top_10_run_scoreres.values()))
plt.xlabel("batsmen")
plt.title("top 10 run scoreres")
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x(), yval + .005, yval)
plt.show()
```

```python
#top 10 bowlers with high madiens(in overs)

def madiens(df,bo_ba,name):
    x = df[(df[bo_ba] == name) & 
           (df['runs_off_bat'] == 0) & 
           (df['extras'] == 0) & 
           (df['noballs'].isna()) & 
           (df['byes'].isna()) & 
           (df['legbyes'].isna())]
    
    return x.shape[0]

m = {}
for i in bowlers:
    m[i] = madiens(ipl_df, 'bowler',i)//6
madiens_top10 = dict(sorted(m.items(), key = lambda x: x[1], reverse = True)[:10])

fig = plt.figure(figsize = (18, 5))
bars = plt.bar(list(madiens_top10.keys()),list(madiens_top10.values()))
plt.xlabel("bowlers")
plt.title("Top 10 bowlers with high madiens(in overs)")
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x(), yval + .005, yval)
plt.show()
```

```python
#average number of runs/wickets per season
runs, wickets = {},{}

for season in seasons:
    season_data = ipl_df[ipl_df['season']==season]
    match_id = list(season_data['match_id'].unique())
    t_mt = len(match_id)
    avg_runs,avg_wt = 0,0
    
    for match in match_id:
        match_data = season_data[season_data['match_id']==match]
        
        r = match_data['runs_off_bat'].sum() + match_data['extras'].sum()
        avg_runs+=r
        
        match_data = season_data[(season_data['player_dismissed'].notna()) &
                                 (season_data['wicket_type'] != 'run out') &  
                                 (season_data['wicket_type'] != 'retired hurt') & 
                                 (season_data['wicket_type'] !='hit wicket') & 
                                 (season_data['wicket_type'] !='obstructing the field')]
        avg_wt = match_data.shape[0]
        
    avg_runs /= t_mt
    avg_wt /= t_mt
    runs[season] = round(avg_runs)
    wickets[season] = avg_wt

del runs[2021]
del wickets[2021]
```
